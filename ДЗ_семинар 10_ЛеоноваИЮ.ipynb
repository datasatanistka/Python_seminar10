```python
import random
import pandas as pd
```

```python
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
data_temp = data.copy()
```

```python
unique_values = data['whoAmI'].unique()
print(unique_values)
```

```python
for value in unique_values:
  data_temp.loc[data_temp['whoAmI']==value,value] = '1'
  data_temp.loc[data_temp['whoAmI']!=value,value] = '0'
```

```python
data_one_hot = data_temp.drop(columns=['whoAmI'])
```
